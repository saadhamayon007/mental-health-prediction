##### Handling CGPA column

```python
df["CGPA"].value_counts().sort_index(ascending=False).head(10)
```

- This shows that cgpa column is on 10 point scale.
- We need it to scale to 04 point scale.

```python
def convert_to_4scale(score):
    if score >= 8.5:
        return 4.0
    elif score >= 8.0:
        return 3.7
    elif score >= 7.5:
        return 3.3
    elif score >= 7.0:
        return 3.0
    elif score >= 6.5:
        return 2.7
    elif score >= 6.0:
        return 2.3
    elif score >= 5.5:
        return 2.0
    elif score >= 5.0:
        return 1.7
    elif score >= 4.5:
        return 1.3
    else:
        return 0.0

df["CGPA_4scale"] = df["CGPA"].apply(convert_to_4scale)
```

```python
# Verifying CGPA scale

print("Value of counts", df["CGPA_4scale"].value_counts())

print("\033[1mSample of original vs converted:\033[0m")
print(df[["CGPA", "CGPA_4scale"]].head(10))
```

##### Sleep Categories

```python
def categorize_sleep(hours):
    if hours < 5:
        return "Insufficient (<5h)"
    elif hours < 6:
        return "Below Recommended (6-7h)"
    elif hours <= 8:
        return "Recommended (7-9h)"
    elif hours <= 9:
        return "Excellent"
    else:
        return "Excessive (>9h)"


df["sleep_category"] = df["Sleep_duration"].apply(categorize_sleep)

print("Sleep Duration Categories : ")
print(df["sleep_category"].value_counts())
```

```python

def group_profession(prof):
    prof = str(prof).lower()
    
    # Student categories
    if 'student' in prof or 'study' in prof:
        return 'Student'
    
    # Healthcare
    elif any(word in prof for word in ['doctor', 'nurse', 'medical', 'health', 'pharma']):
        return 'Healthcare'
    
    # Tech/IT
    elif any(word in prof for word in ['engineer', 'developer', 'software', 'it', 'tech', 'programmer']):
        return 'Technology'
    
    # Business
    elif any(word in prof for word in ['manager', 'business', 'admin', 'executive', 'analyst']):
        return 'Business'
    
    # Education
    elif any(word in prof for word in ['teacher', 'professor', 'lecturer', 'education']):
        return 'Education'
    
    # Creative
    elif any(word in prof for word in ['designer', 'artist', 'writer', 'creative']):
        return 'Creative'
    
    else:
        return 'Other'

# Apply grouping
df['Profession_grouped'] = df['Profession'].apply(group_profession)

# One-hot encode the grouped version
prof_dummies = pd.get_dummies(df['Profession_grouped'], prefix='Profession', drop_first=True)
df = pd.concat([df, prof_dummies], axis=1)

# ✅ DROP THE ORIGINAL COLUMNS
df = df.drop(['Profession', 'Profession_grouped'], axis=1)

# Verify
print("Columns after dropping:", [col for col in df.columns if 'Profession' in col])
```

```python
# Convert existing True/False columns to 1/0
profession_cols = [col for col in df.columns if 'Profession_' in col]

for col in profession_cols:
    df[col] = df[col].astype(int)  # True→1, False→0

# Or using replace
for col in profession_cols:
    df[col] = df[col].replace({True: 1, False: 0})

print("Converted to 1/0:")
print(df[profession_cols].head())
```
